```
import pandas as pd

df = pd.read_csv('census.csv')
df.head()
```

**Keep rows with specific value**

- Where function

```
data = df.where(df['SUMLEV']==50).dropna().set_index(['STNAME','CTYNAME'])
data
```

**Rename Column**

```
data = data.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
data
```

**Find Min/Max from Multiple Column**
 ApplyMap

```
import numpy as np
```

```
def min_max(row):
    dataF = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(dataF)
    row['min'] = np.min(dataF)
    return row
```

```
data = data.apply(min_max, axis=1)
data.head()
```

# **GroupBy**

```
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + ' have an average population of ' + str(avg))
```

```
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + ' have an average population of ' + str(avg))
```

**Agg Function**

```
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
```

```
data = df.groupby('STNAME').agg({'CENSUS2010POP': np.average})
data.head()
```

**Multiple agg function on Single Column**

```
data = (df.set_index('STNAME').groupby(level=0)['CENSUS2010POP']
          .agg([('Mean',np.average),('Sum',np.sum)]))
data.head()
```

**Multiple agg function on Multiple Column**

```
data = (df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']
          .agg([('Mean',np.average),('Sum',np.sum)]))
data.head()
```
